```python
import pandas as pd

df = pd.read_csv("Dummy_Sales_Data_v1.csv")
```

# head

```python
#This function helps you to get the first few rows of the dataset. By default, it returns the first 5 rows.
#However, you can change this number by simply mentioning the desired number of rows.

df.head()
```

# tail


```python
#This function helps you to get the last few rows of the dataset. By default, it returns the last
#5 rows, and similar to .head(), you can simply mention the desired number of rows.

df.tail()
```

# sample

```python
#This function is used to get a randomly selected row, column, or both from a dataset.

df.sample()
```

# info


```python
#This function returns a quick summary of the DataFrame. This includes information about
#column names and their respective data types, missing values, and memory consumption.

df.info()
```

# describe

```python
# This function returns descriptive statistics about the data. This includes minimum,
# maximum, mean (central tendency), standard deviation (dispersion) of the values in
# numerical columns, and the count of all non-null values in the data.

df.describe()
```

# Query

```python
# This function is used to query the DataFrame based on an expression. An expression can
# be as simple as a single condition and as complex as a combination of multiple conditions.

df.query("Quantity>95")
```

# location

```python
#This function returns the group of rows and columns
#identified by their labels or names.

df.loc[100,'Sales_Manager']
```

# unique

```python
#This function returns the list of unique values in a column or series. Instead of applying on
#complete dataFrame it works only on the selected single column.

df["Sales_Manager"].unique()
```

# nunique

```python
#It is used to calculate the number of unique values in each column of a DataFrame. 
#It returns count of distinct values for each column.

result=df.nunique()
print(result)
```

# isNull

```python
#It is used to check for missing or null values in a DataFrame. 
#It returns a boolean value showing whether each element in the DataFrame is null or not.
df.isnull()
```

# fillna

```python
#The fillna() method in pandas is used to fill missing or null values
#in a DataFrame with specified values or strategies.

a = df.fillna("product")
print(a)
```

# sort_values

```python
#It is used to sort a DataFrame by one or more columns. 
#It allows you to specify the column(s) to sort by and the order of sorting 

a = df.sort_values("Quantity")
print(a)
```

# values count

```python
#It is used count the occurrences of unique values in a column of a DataFrame.

b = df.value_counts("Sales_Manager")
print(b)
```

# nLargest

```python
#It is used to retrieve the rows with the 
#largest values in one or more columns of a DataFrame.

c = df.nlargest(4,"UnitPrice(USD)")
print(c)
```

# nSmallest

```python
#It is used to retrieve the rows with the smallest values in one or more columns of a DataFrame. 

d = df.nsmallest(3,"UnitPrice(USD)")
print(d)
```

# copy

```python
#It is used to create a deep copy of a DataFrame. 
#It allows you to create a new DataFrame object that is independent of the original DataFrame.

a = df.copy()
print(a)
```

# rename

```python
#It is used to rename columns or index labels of a DataFrame. 
#It allows you to change the names of one or more columns or

rename= df.rename(columns={'Quantity': 'Quan'})
print(rename)
```

# where

```python
#It is used to replace values in a DataFrame based on a specified condition. 

import numpy as np
where =  df.where(df['Quantity'] < 92, np.nan)
print(where)
```

# drop

```python
#It is used to remove specified rows or columns from a DataFrame. 
#It allows you to delete rows or columns by specifying their labels or positions
drop=df.drop("Quantity",axis=1)
print(drop)
```

# corellation

```python
#It is used to compute the correlation between columns in a DataFrame. 
#It calculates the pairwise correlation between numeric columns and returns a correlation matrix.
df.corr()
```
